## Batteries Calculation

```python
# Function to calculate joltages
def joltage_calculation(data):
    # List to store joltages
    joltage_list = []

    # Iterate through each bank
    for bank in data['Bank']:
        # Dictionary to store results
        dicti = {}
        # Flag for exception handling
        flag = False
        # Convert bank to list
        bank_list = [int(digit) for digit in str(bank)]
        # Find required digits
        a = max(bank_list)
        b = bank_list.index(a)
        bank_list[b] = 0
        c = max(bank_list)
        d = bank_list.index(c)
        bank_list[d] = 0
        dicti[b] = a
        dicti[d] = c
        # Check for exception
        if dicti[min(dicti.keys())] < dicti[max(dicti.keys())]:
            if max(dicti.keys()) < len(bank_list) - 1:
                flag = True
                e = max(bank_list[max(dicti.keys()) + 1:])
        # Create final joltage for the bank
        if flag == False:
            final_joltage = int(str(dicti[min(dicti.keys())]) + str(dicti[max(dicti.keys())]))
        else:
            final_joltage = int(str(dicti[max(dicti.keys())]) + str(e))
        joltage_list.append(final_joltage)
    return joltage_list
```

```python
# Calculate joltafges for test data
print(sum(joltage_calculation(test_data)))
```

```python
# Calculate joltages for inut data
print(sum(joltage_calculation(input_data)))
```
