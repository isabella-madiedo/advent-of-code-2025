## Batteries Calculation - Method 1

```python
# Function to calculate joltages
def joltage_calculation(data):
    # List to store joltages
    joltage_list = []

    # Iterate through each bank
    for bank in data['Bank']:
        # Dictionary to store results
        dicti = {}
        # Flag for exception handling
        flag = False
        # Convert bank to list
        bank_list = [int(digit) for digit in str(bank)]
        # Find required digits
        a = max(bank_list)
        b = bank_list.index(a)
        bank_list[b] = 0
        c = max(bank_list)
        d = bank_list.index(c)
        bank_list[d] = 0
        dicti[b] = a
        dicti[d] = c
        # Check for exception
        if dicti[min(dicti.keys())] < dicti[max(dicti.keys())]:
            if max(dicti.keys()) < len(bank_list) - 1:
                flag = True
                e = max(bank_list[max(dicti.keys()) + 1:])
        # Create final joltage for the bank
        if flag == False:
            final_joltage = int(str(dicti[min(dicti.keys())]) + str(dicti[max(dicti.keys())]))
        else:
            final_joltage = int(str(dicti[max(dicti.keys())]) + str(e))
        joltage_list.append(final_joltage)
    return joltage_list
```

```python
# Calculate joltages for test data
print(f'The total joltage is {sum(joltage_calculation(test_data))}')
```

```python
# Calculate joltages for inut data
print(f'The total joltage is {sum(joltage_calculation(input_data))}')
```

## Batteries Calculation - Method 2

```python
# Function 2 to calculate joltages
def joltage_calculation_v2(data):
    # Iterate through each bank
    digits = 12 # Joltage digits per bank
    joltage_list = [] # Store for joltages
    
    for bank in data['Bank']:
        bank_joltage_digits = '' # Store for bank joltage digits
        start = 0 # Start index
        bank_list = [int(digit) for digit in str(bank)] # Convert bank to list
        bank_len = len(bank_list)
        end = bank_len - digits + 1 # End index

        while len(bank_joltage_digits) < digits:
            maxi = max(bank_list[start:end]) # Iterate sequentially to find the max digit within each range [start:end]
            index = bank_list[start:end].index(maxi) + start # Find index for maz digit within the range
            bank_joltage_digits += str(maxi)
            start = index + 1
            end += 1
        joltage_list.append(int(bank_joltage_digits))
    return joltage_list
```

```python
# Calculate new joltage for test data
print(f'The new total voltage is {sum(joltage_calculation_v2(test_data))}')
```

```python
# Calculate new joltage for input data
print(f'The new total voltage is {sum(joltage_calculation_v2(input_data))}')
```
