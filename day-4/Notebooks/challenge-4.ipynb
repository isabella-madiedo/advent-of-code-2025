## Libraries

```python
import pandas as pd
```

## Data

```python
# Load test data
test_data = pd.read_csv('../Data/test-data.csv', header=None, names=['patterns'])

# Load input data
input_data = pd.read_csv('../Data/input-data.csv', header=None, names=['patterns'])
```

```python
# Separate each column and row of paperl rolls
data = []
for row in input_data['patterns']:
    data.append(list(row))
```

```python
# Transform the input into binary values
for i in range(len(data)):
    for j in range(len(data[0])):
        if data[i][j] == '@':
            data[i][j] = 1
        else:
            data[i][j] = 0
```

## Finding Rolls of Christmas Paper

```python
# Find the paper roles which can be accessed
cont = 0
for i in range(len(data)):
    for j in range(len(data[0])):
        if data[i][j] == 0:
            continue

        lista = []
        if i == 0 or j == 0 or i == len(data)-1 or j == len(data[0])-1:
            if i > 0 and j > 0:
                lista.append(data[i-1][j-1])
            else:
                lista.append(0)
            if i > 0:
                lista.append(data[i-1][j])
            else:
                lista.append(0)
            if i > 0 and j < len(data[0])-1:
                lista.append(data[i-1][j+1])
            else:
                lista.append(0)
            if j > 0:
                lista.append(data[i][j-1])
            else:
                lista.append(0)
            if j < len(data[0])-1:
                lista.append(data[i][j+1])
            else:
                lista.append(0)
            if i < len(data)-1 and j > 0:
                lista.append(data[i+1][j-1])
            else:
                lista.append(0)
            if i < len(data)-1:
                lista.append(data[i+1][j])
            else:
                lista.append(0)
            if i < len(data)-1 and j < len(data[0])-1:
                lista.append(data[i+1][j+1])
            else:
                lista.append(0)
        else:
            lista.append(data[i-1][j-1])
            lista.append(data[i-1][j])
            lista.append(data[i-1][j+1])
            lista.append(data[i][j-1])
            lista.append(data[i][j+1])
            lista.append(data[i+1][j-1])
            lista.append(data[i+1][j])
            lista.append(data[i+1][j+1])
        if sum(lista) < 4:
            cont += 1
```

```python
print(f'The total number of paper rolls which can be accessed is {cont}')
```

## Removing Rolls of Christmas Paper

```python
# Find the paper roles which can be accessed
cont = 0
rolls = []
flag = False

while flag == False:
    cont_2 = 0
    for i in range(len(data)):
        for j in range(len(data[0])):
            if data[i][j] == 0:
                continue

            lista = []
            if i == 0 or j == 0 or i == len(data)-1 or j == len(data[0])-1:
                if i > 0 and j > 0:
                    lista.append(data[i-1][j-1])
                else:
                    lista.append(0)
                if i > 0:
                    lista.append(data[i-1][j])
                else:
                    lista.append(0)
                if i > 0 and j < len(data[0])-1:
                    lista.append(data[i-1][j+1])
                else:
                    lista.append(0)
                if j > 0:
                    lista.append(data[i][j-1])
                else:
                    lista.append(0)
                if j < len(data[0])-1:
                    lista.append(data[i][j+1])
                else:
                    lista.append(0)
                if i < len(data)-1 and j > 0:
                    lista.append(data[i+1][j-1])
                else:
                    lista.append(0)
                if i < len(data)-1:
                    lista.append(data[i+1][j])
                else:
                    lista.append(0)
                if i < len(data)-1 and j < len(data[0])-1:
                    lista.append(data[i+1][j+1])
                else:
                    lista.append(0)
            else:
                lista.append(data[i-1][j-1])
                lista.append(data[i-1][j])
                lista.append(data[i-1][j+1])
                lista.append(data[i][j-1])
                lista.append(data[i][j+1])
                lista.append(data[i+1][j-1])
                lista.append(data[i+1][j])
                lista.append(data[i+1][j+1])
            if sum(lista) < 4:
                cont += 1
                cont_2 += 1
                rolls.append([i,j])

    for roll in rolls:
        data[roll[0]][roll[1]] = 0
    
    if cont_2 == 0:
        flag = True
```
