```python
import pandas as pd
```

## Data

```python
# Load test data
test_data = pd.read_csv('../Data/test-data.csv', header=None, names=['Ranges'])

# Load input data
input_data = pd.read_csv('../Data/input-data.csv', header=None, names=['Ranges'])
```

## Find Incorrect IDs

```python
# Function to find incorrect IDs
def find_incorrect_ids(data):
    list_invalid_ids=[]
    for rango in data['Ranges']:
        start = int(rango.split('-')[0])
        end = int(rango.split('-')[1])

        for number in range(start, end + 1):
            length = len(str(number))
            if length % 2 != 0:
                continue
            else:
                part_1 = str(number)[:length // 2]
                part_2 = str(number)[length // 2:]
                if part_1 == part_2:
                    list_invalid_ids.append(number)
    return sum(list_invalid_ids)
```

```python
# Sum of all invalid IDs for test data
find_incorrect_ids(test_data)
```

```python
# Sum of all invalid IDs for input data
find_incorrect_ids(input_data)
```
