## Libraries

```python
import pandas as pd
import math
```

## Data

```python
# Load input data
input_data = pd.read_excel('../Data/input-data.xlsx', header=None, names=['coordinates'])
```

## Circuits

```python
# Calculate all distances and store them in a dictionary for each pair of coordinates
coordinates =list(input_data['coordinates'])
counter = -1
distances = {}
for c_1 in coordinates:
    counter += 1
    for c_2 in coordinates[counter:]:
        if c_1 == c_2:
            continue
        distances[c_1 + '-' + c_2] = math.dist(eval(c_1), eval(c_2))
```

```python
# Form the circuits
circuits = []
for _ in range(1000):
    min_dist = min(distances.values())
    min_cs = [k for k in distances.keys() if distances[k] == min_dist][0]
    distances[min_cs] = float('inf')

    c_1 = min_cs.split('-')[0]
    c_2 = min_cs.split('-')[1]
    circuit_with_c1 = None
    circuit_with_c2 = None
    
    for circuit in circuits:
        if c_1 in circuit:
            circuit_with_c1 = circuit
        if c_2 in circuit:
            circuit_with_c2 = circuit
    
    if circuit_with_c1 is not None and circuit_with_c2 is not None:
        if circuit_with_c1 is not circuit_with_c2:
            circuit_with_c1.update(circuit_with_c2)
            circuits.remove(circuit_with_c2)
    elif circuit_with_c1 is not None:
        circuit_with_c1.add(c_2)
    elif circuit_with_c2 is not None:
        circuit_with_c2.add(c_1)
    else:
        circuits.append({c_1, c_2})
```

```python
# Get the sizes of the three largest circuits
sizes = []
for circuit in circuits:
    sizes.append(len(circuit))
sizes.sort()
total = sizes[-1] * sizes[-2] * sizes[-3]
print(f'The answer to the circuit puzzle is {total}')
```
