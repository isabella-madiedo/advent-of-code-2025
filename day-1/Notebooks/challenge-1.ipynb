## Data

```python
# Test Data Import
test_data = pd.read_csv('../Data/test-data.csv', header=None)
# Real Data Import
real_data = pd.read_csv('../Data/input.csv', header=None)
```

```python
# Test Data Preparation
test_data['Dir'] = test_data[0].str[0]
test_data['Steps'] = test_data[0].str[1:].astype(int)

# Real Data Preparation
real_data['Dir'] = real_data[0].str[0]
real_data['Steps'] = real_data[0].str[1:].astype(int)
```

## Password Calculation - Method 1

```python
# Password calculation function
def pass_calc(data):
    dial = 50
    counter = 0

    for i in range(len(data)):
        if dial == 0:
            counter += 1
        if data.loc[i, 'Dir'] == 'L':
            dial = (dial - data.loc[i, 'Steps']) % 100
        else:
            dial = (dial + data.loc[i, 'Steps']) % 100
    return counter, dial
```

```python
# Password calculation results
print(f'The password for the test data is: {pass_calc(test_data)[0]} and the dial end position is: {pass_calc(test_data)[1]}')  # Test data
print(f'The password for the real data is: {pass_calc(real_data)[0]} and the dial end position is: {pass_calc(real_data)[1]}')   # Real data
```

## Password Calculation - Method 2

```python
# Password calculation function
def pass_calc_2(data):
    # counts movements ending at 0
    counter_end = pass_calc(data)[0]
    
    # counts movements passing through 0
    dial = 50
    counter = 0

    for i in range(len(data)):
        if data.loc[i, 'Dir'] == 'R':
            counter = counter + (data.loc[i, 'Steps'] // 100)
            if (data.loc[i, 'Steps'] % 100) + dial > 100:
                counter += 1
            dial = (dial + data.loc[i, 'Steps']) % 100
        else:
            counter = counter + (data.loc[i, 'Steps'] // 100)
            if dial - (data.loc[i, 'Steps'] % 100) < 0:
                counter += 1
            counter = counter - 1 if dial == 0 else counter
            dial = (dial - data.loc[i, 'Steps']) % 100 
    return counter_end + counter, dial
```

```python
# Password calculation 2 results
print(f'The password for the test data is: {pass_calc_2(test_data)[0]} and the dial end position is: {pass_calc_2(test_data)[1]}')  # Test data
print(f'The password for the real data is: {pass_calc_2(real_data)[0]} and the dial end position is: {pass_calc_2(real_data)[1]}')   # Real data
```
