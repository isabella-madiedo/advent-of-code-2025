## Libraries

```python
import pandas as pd
```

## Data

```python
# Load test data
test_rangos = pd.read_csv('../Data/test-data-ranges.csv', header=None, names=['rango'])
test_ingredients = pd.read_csv('../Data/test-data-ingredients.csv', header=None, names=['ingredient'])

# Load input data
input_rangos = pd.read_csv('../Data/input-data-ranges.csv', header=None, names=['rango'])
input_ingredients = pd.read_csv('../Data/input-data-ingredients.csv', header=None, names=['ingredient'])
```

## Find Fresh Ingredients

```python
def find_fresh_ingredients(ingredients, rangos):
    fresh_ingredients = 0
    for ingredient in ingredients:
        for rango in rangos:
            start = int(rango.split('-')[0])
            end = int(rango.split('-')[1])
            if ingredient >= start and ingredient <= end:
                fresh_ingredients += 1
                break
    return fresh_ingredients
```

```python
# Number of fresh ingrdients in test data
print(f'The number of fresh ingredients in the test data is {find_fresh_ingredients(test_ingredients["ingredient"], test_rangos["rango"])}')
```

```python
# Number of fresh ingrdients in input data
print(f'The number of fresh ingredients in the test data is {find_fresh_ingredients(input_ingredients["ingredient"], input_rangos["rango"])}')
```
