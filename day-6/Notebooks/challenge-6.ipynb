## Math Homework

```python
# Function to calculate each math homework problem
def math_hw(data):
    total_op = []
    for row in range(len(data)):
        if data.iloc[row, 4] == '+':
            total_op.append(data.iloc[row, 0] + data.iloc[row, 1] + data.iloc[row, 2] + data.iloc[row, 3])
        else:
            total_op.append(data.iloc[row, 0] * data.iloc[row, 1] * data.iloc[row, 2] * data.iloc[row, 3])
    return sum(total_op)
```

```python
# Input data results
math_hw(input_data)
```
